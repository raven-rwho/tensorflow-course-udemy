# Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(0,10)
```

```python
y = x**2
```

check the documentation here to fiddle around with the visualization that is offered by the plot method

```python
plt.plot(x,y,'r--')
plt.xlim(0,4)
plt.ylim(0,10)
plt.title('Title')
plt.xlabel('x label')
plt.ylabel('y label')
```

```python
mat = np.arange(0,100).reshape(10,10)
```
